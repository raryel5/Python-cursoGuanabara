```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import plotly.express as px

tabela0209 = pd.read_csv("fm0209.csv")

# excluir alguma coluna, posicao 1, por exemplo
# tabela = tabela.drop("nomeColuna", axis=1)

# mudar o nome da coluna de campo eletrico
tabela0209.rename(columns={'Electric_Field_V_m': 'Field'}, inplace=True)

display(tabela0209)
```

```python
#type(tabela0209)
temp = tabela0209['Field'].tolist()
campo = pd.Series(temp)
# display(campo)

# print(campo[4])
type(campo)
```

```python

variacao = []

for i in range(len(campo)):
    if i == 0:
        variacao.append(0)
    else:
        variacao.append(campo[i] - campo[i-1])

x = np.arange(0, len(variacao), 1)
plt.scatter(x, variacao)
plt.show()
# display(variacao)
# variacao.append(1)
# print(variacao[0])
```
